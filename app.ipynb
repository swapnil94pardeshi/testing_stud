```python
import pandas as pd
df=pd.read_csv("student_risk_dataset.csv")

df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Attendance_Percentage'].fillna(df['Attendance_Percentage'].median(),inplace=True)

df['Homework_Completion (%)'].fillna(df['Homework_Completion (%)'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df[(df["Attendance_Percentage"]>=0) & (df['Attendance_Percentage']<=100)]
```

```python
df=df[(df["Homework_Completion (%)"]>=0) & (df['Homework_Completion (%)']<=100)]
```

```python
score_cols=['Maths_Score','Science_Score','English_Score']

for col in score_cols:
    df=df[(df[col]>=0) & (df[col]<=100)]
```

```python
df=df.drop_duplicates(subset='Student_ID')
```

```python
df['Parent_Involvement']=df['Parent_Involvement'].str.strip().str.lower()

df["Parent_Involvement"]=df['Parent_Involvement'].replace({'yes':'Yes',
'y':'Yes',
'no':'No',
'n':'No'})
```

```python
df['Region']=df['Region'].str.title()
df['Fee_Status']=df['Fee_Status'].str.title()
df['Payment_Type']=df['Payment_Type'].str.title()
```

```python
df[(df['Overall_Percentage']>80) & (df['Attendance_Percentage']<40)]
```

```python
def improved_risk(row):
    if row['Overall_Percentage']<40 or row['Attendance_Percentage']<50 or row['Fee_Status']=='Overdue':
        return 'High'
    elif row['Overall_Percentage']<60:
        return 'Medium'
    else:
        return 'Low'
df['Risk_Level']=df.apply(improved_risk,axis=1)
```

```python
df.to_csv("cleaned_student_dataset.csv",index=False)
```
