```python
import pandas as pd
import random
from datetime import datetime, timedelta

# -----------------------------
# CONFIG
# -----------------------------
NUM_STUDENTS = 10000
TESTS_PER_STUDENT = 3

subjects = ["Maths"]
topics = ["Algebra", "Geometry", "Trigonometry", "Mensuration", "Statistics"]

# -----------------------------
# 1. STUDENTS DATA
# -----------------------------
students = []

for i in range(1, NUM_STUDENTS + 1):
    students.append({
        "student_id": i,
        "name": f"Student_{i}",
        "class": random.choice([9, 10]),
        "board": random.choice(["CBSE", "SSC"])
    })

students_df = pd.DataFrame(students)

# -----------------------------
# 2. TESTS DATA
# -----------------------------
tests = []
test_id = 1

for student in students:
    for _ in range(TESTS_PER_STUDENT):
        total_marks = 100
        obtained_marks = random.randint(30, 95)
        time_taken = random.randint(30, 90)

        test_date = datetime.now() - timedelta(days=random.randint(0, 30))

        tests.append({
            "test_id": test_id,
            "student_id": student["student_id"],
            "subject": "Maths",
            "total_marks": total_marks,
            "obtained_marks": obtained_marks,
            "time_taken": time_taken,
            "test_date": test_date.strftime("%Y-%m-%d")
        })

        test_id += 1

tests_df = pd.DataFrame(tests)

# -----------------------------
# 3. TOPIC PERFORMANCE DATA
# -----------------------------
topic_data = []
topic_id = 1

for test in tests:
    for topic in topics:
        max_marks = 20
        obtained_marks = random.randint(5, 20)

        topic_data.append({
            "id": topic_id,
            "test_id": test["test_id"],
            "topic": topic,
            "max_marks": max_marks,
            "obtained_marks": obtained_marks
        })

        topic_id += 1

topic_df = pd.DataFrame(topic_data)

# -----------------------------
# SAVE FILES
# -----------------------------
students_df.to_csv("students.csv", index=False)
tests_df.to_csv("tests.csv", index=False)
topic_df.to_csv("topic_performance.csv", index=False)

print("✅ Data Generated Successfully!")
```

# Load Your Data


```python
import pandas as pd

students=pd.read_csv("students.csv")
tests=pd.read_csv("tests.csv")
topics=pd.read_csv("topic_performance.csv")
```

## Merge Data


```python
data=topics.merge(tests,on="test_id").merge(students,on="student_id")
```

```python
data.head(20)
```

## Calculate Accuracy


```python
data["accuracy"]=(data["obtained_marks_x"]/data["max_marks"]) *100
```

```python
data.head(10)
```

# Find Topic Performance Per Student


```python
topic_summary=data.groupby(["student_id","topic"])["accuracy"].mean().reset_index()
```

```python
data.head()
```

# Identify Weak & Strong Topics


```python
def classify_topic(acc):
    if acc<50:
        return"Weak"
    elif acc>75:
        return"Strong"
    else:
        return "Average"

topic_summary["level"]=topic_summary["accuracy"].apply(classify_topic)
```

# Identify Slow Students


```python
time_summary=tests.groupby("student_id")["time_taken"].mean().reset_index()

def classify_speed(time):
    if time > 70:
        return "slow"
    elif time<40:
        return "fast"
    else:
        return "Normal"
    
time_summary["speed"]=time_summary["time_taken"].apply(classify_speed)
```

# Recommendation Engine


```python
def recommend(student_id):
    student_topics = topic_summary[topic_summary["student_id"] == student_id]
    student_speed = time_summary[time_summary["student_id"] == student_id]["speed"].values[0]

    weak = student_topics[student_topics["level"] == "Weak"]["topic"].tolist()
    strong = student_topics[student_topics["level"] == "Strong"]["topic"].tolist()

    print(f"\n📌 Recommendations for Student {student_id}:\n")

    if weak:
        print("🔴 Focus on weak topics:", ", ".join(weak))
        print("👉 Practice basic questions and revise concepts")

    if strong:
        print("\n🟢 Strong topics:", ", ".join(strong))
        print("👉 Try advanced questions")

    if student_speed == "Slow":
        print("\n⏱️ Work on speed: Take timed tests")

    elif student_speed == "Fast":
        print("\n⚡ Good speed: Try accuracy improvement")

    print("\n📚 Suggested Action: Attempt next test tomorrow")
```

```python
recommend(10)
```
